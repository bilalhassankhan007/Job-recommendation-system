```python
import nltk 
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer() ##storing object for Porter
```

```python
def cleaning(txt):
    cleaned_txt = re.sub(r'[^a-zA-Z0-9\s]','',txt)
    tokens = word_tokenize(cleaned_txt.lower())
    stemming = [ps.stem(word) for word in tokens if word not in stopwords.words('english')]
    return " ".join(stemming)
```

```python
df['title'] = df['title'].apply(lambda x:cleaning(x)) 
df['description'] = df['description'].apply(lambda x:cleaning(x))
```

```python
df['new_col'] = df['title']+ " " + df['description'] ##adding two columns into one and storing  in a variable 
```

### Vectorization 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfif = TfidfVectorizer()
matrix = tfif.fit_transform(df['new_col'])
similarity = cosine_similarity(matrix) ##extracting the similar words from the "new_col"
```

```python
similarity ##vectors formed with respect to the similar items
```

```python
def recommendation(job_title):
    idx = df[df['title'] == job_title].index[0]
    idx = df.index.get_loc(idx)
    distances = sorted(list(enumerate(similarity[idx])), key=lambda x:x[1], reverse=True)[1:]
    
    for i in distances:
        print(df.iloc[i[0]].title) 
```
